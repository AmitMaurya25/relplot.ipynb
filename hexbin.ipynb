```python
import pandas as pd
import numpy as np
df= pd.read_csv("nhanes_2015_2016.csv")
df.head()
```

```python
df.columns
```

```python
df.plot.hexbin(x = 'BMXARMC', y='BMXLEG', gridsize = 20)
```

```python
df.plot.hexbin(x = 'BMXARMC', y='BMXLEG', 
               C = 'BMXHT',
               reduce_C_function=np.max,
               gridsize = 15,
              figsize=(8, 6))
```

```python
df.plot.hexbin(x = 'BMXARMC', y='BMXLEG', 
               C = 'BMXHT',
               reduce_C_function=np.max,
               gridsize = 15,
              figsize=(8, 6),
              cmap = 'viridis')
```

```python
df.plot.hexbin(x = 'BMXARMC', y='BMXLEG', 
               C = 'BMXHT',
               reduce_C_function=np.max,
               gridsize = 15,
               edgecolors = 'face',
              figsize=(8, 6),
              cmap = 'viridis')
```

```python
df.plot.hexbin(x = 'BMXARMC', y='BMXLEG', 
               C = 'BMXHT',
               reduce_C_function=np.max,
               gridsize = 20,
               edgecolors = 'face',
               mincnt = 10,
              figsize=(8, 6),
              cmap = 'viridis')
```
