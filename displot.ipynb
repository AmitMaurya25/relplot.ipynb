```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.displot(tips, x = "total_bill")
```

```python
sns.displot(tips, x = "total_bill", kde=True, rug = True)
```

```python
sns.displot(tips, x = "total_bill", kind = 'kde')
```

```python
sns.displot(tips, x = "total_bill", kind = 'ecdf')
```

```python
sns.displot(tips, x = "total_bill",y = 'tip')
```

```python
sns.displot(tips, x = "total_bill",y = 'tip', kind='kde')
```

```python
sns.displot(tips, x = "total_bill", hue = "day", multiple= 'stack')
```

```python
sns.displot(tips, x = "total_bill", kind = 'kde', hue = "day", multiple= 'stack')
```

```python
sns.displot(tips, x = "total_bill", kind = 'kde', hue = "day")
```

```python
sns.displot(tips, x = "total_bill", hue = "day", multiple= 'dodge', bins = 5, shrink = 0.8)
```

```python
sns.displot(tips, x = "total_bill", hue = "day", col='day',
           height=5, aspect=0.6)
```

```python
sns.displot(tips, x = "total_bill", hue = "day", col='day', row = "time",
           height=5, aspect=0.6)
```
