```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid")
t = sns.load_dataset("titanic")
t
```

```python
plt.figure(figsize = (8, 6))
sns.stripplot(x = "embark_town", y = "age", data = t, jitter = 0.15)
```

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(x = "embark_town", y = "age", data = t)
```

```python
t['embark_town'].unique()
```

```python
plt.figure(figsize = (8, 6))
sns.stripplot(x = "embark_town", y = "age", data = t, jitter = 0.15, hue = "sex",
              order = ['Queenstown', 'Southampton', 'Cherbourg']
             )
```

```python
plt.figure(figsize = (8, 6))
sns.stripplot(x = "embark_town", y = "age", data = t, jitter = 0.15, palette = "Set2",
              linewidth = 1, edgecolor = 'grey', marker = 'd', size = 10, alpha = 0.5
             )
```

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = "embark_town", y = "age", data = t)
sns.stripplot(x = "embark_town", y = "age", data = t, color = "0.3", alpha = 0.8)
```

```python
plt.figure(figsize = (8, 6))
sns.violinplot(x = "embark_town", y = "age", data = t)
sns.stripplot(x = "embark_town", y = "age", data = t, color = "0.3", alpha = 0.8)
```

```python
sns.catplot(x = "embark_town", y = "age", data = t,
            kind = "strip", col = 'pclass', row = "alive", hue = "sex",
            height = 5, aspect = 0.7
           )
```
