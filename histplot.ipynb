```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, y = 'total_bill')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', binwidth = 5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             binwidth = 5, fill = False, element = "step",
             hue = 'time'
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             fill = False, element = "step",
             hue = 'time', log_scale = True
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             hue = 'time', kde = True, multiple = "stack"
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             hue = 'time', kde = True, multiple = "stack"
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             hue = 'time', kde = True, stat='probability'
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             hue = 'day', element='step', fill = False, cumulative = True
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', 
             hue = 'day', element='step', hue_order = ['Sat', "Sun"],
             fill = False, cumulative = True
            )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'day', shrink = 0.8)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'day', hue = 'time',
             shrink = 0.8)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'day', hue = 'time', multiple = 'dodge',
             shrink = 0.8)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', alpha = 0.6)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = tips, x = 'total_bill', y = 'day', hue = 'time', alpha = 0.6)
plt.show()
```
