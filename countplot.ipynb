```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
```

```python
p = sns.load_dataset("penguins")
```

```python
p
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = "species", data = p)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = "species", data = p, palette = "Set2")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = "species", data = p,
              facecolor = (0, 0, 0, 0),
              linewidth = 3, edgecolor = sns.color_palette('dark', 3)
             )
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = "species", data = p,hue = "sex"
             )
plt.show()
```

```python
sns.catplot(x = "species", data = p,hue = "sex",
              col = "island", kind = "count",
            height = 5, aspect = 0.8
             )
```
