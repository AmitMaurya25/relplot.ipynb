```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style = "whitegrid")
tips = sns.load_dataset("tips")
tips
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, ci = 75)
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, ci = None)
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, ci = 'sd')
```

```python
sns.barplot(y = "day", x = "total_bill", data = tips, ci = 'sd')
```

```python
from numpy import median 
sns.barplot(x = "day", y = "total_bill", data = tips, estimator = median)
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, hue = "sex")
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, 
            capsize = 0.1, color = "red", saturation = 0.4)
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, 
            capsize = 0.1, color = "red", saturation = 0.4, 
            edgecolor = '0.3', linewidth = 3
           )
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, 
            capsize = 0.1, facecolor = (0, 0, 0, 0), 
            edgecolor = '0.3', linewidth = 3
           )
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, hue = "sex",
            kind = "bar", col = "size", col_wrap = 2, capsize = 0.1
           )
```
